```python
import neuro_lib    # Импорт своей библиотеки для работы с Neuro обработкой файлов
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
def file_plot_graph (file, V_const = 5.0, R_dop = 10, Name = "None"): # вывод графиков всех результатов
    neuro_lib.replace (file)
    neuro_lib.replace (file, '         ', '\t')
    data = pd.read_csv(file, header=None, sep = "\t")
    
    
    gridsize = (14, 1)
    fig = plt.figure(figsize=(12, 6))
    plot_signal = plt.subplot2grid(gridsize, (0, 0), rowspan=6)
    plot_led = plt.subplot2grid(gridsize, (6, 0))
    plot_mosfet = plt.subplot2grid(gridsize, (7, 0), rowspan=3)
    plot_r_t = plt.subplot2grid(gridsize, (10, 0), rowspan=3)
    plot_dop = plt.subplot2grid(gridsize, (13, 0))
    
    # Signal
    plot_signal.set_title(Name)
    cond = 1/((V_const - data[3])*R_dop/data[3])
    plot_signal.scatter(data[0]/1000, cond*1000000, s=2, label="Signal (mkS)", color='g')
    #plot_signal.set_xlabel('Time (sec)')
    #plot_signal.set_ylabel('')
    plot_signal.legend(loc = 1)
    
    # LED
    plot_led.plot(data[0]/1000, data[4], label="LED (bin)", color='m')
    plot_led.set_ylim(-0.2, 1.2)
    plot_led.legend(loc = 1)
    
    # MOSFET
    plot_mosfet.plot(data[0]/1000, data[5], label="MOSFET (V)", color='grey')
    plot_mosfet.set_ylim(-0.4, 5.4)
    plot_mosfet.legend(loc = 1)
    
    # R_heat
    plot_r_t.plot(data[0]/1000, data[2]/data[1], label="R_Heat (Om)", color='r')
    plot_r_t.set_ylim(-4, 204)
    plot_r_t.legend(loc = 1)
    
    # Bin_dop
    plot_dop.plot(data[0]/1000, data[6], label="Dop (bin)", color='k')
    plot_dop.set_ylim(-0.2, 1.2)
    plot_dop.legend(loc = 1)
```

```python
def table_empty_str (file): # функция вывода строк вокруг пробельной строки (строки переключения)
    file_open = open(file, "r")
    buf = 0
    while True:
        line = file_open.readline()
        if (line == '\n'):
            print (buf)
            line = file_open.readline()
            print (line, "............................................")
        if not line:
            break
        buf = line
    file_open.close
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_one_impuls(file, s_start = 162000, s_end = 248000, flag = 1): # график части в скобках
    data = pd.read_csv(file, header=None, sep = "\t")
    data_new = data[(data[0] > s_start*1000) & (data[0] < s_end*1000)]
    data_new[0] = data_new[0] - s_start*1000
    if flag > 0:
        plt.figure(figsize=(9, 4))
        plt.plot(data_new[0]/1000, data_new[3])
    return (data_new)
```
